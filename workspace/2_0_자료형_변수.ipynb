### 응용

* 이를 활용하면 서로 값을 바꾸고 싶은 경우 아래와 같이 활용 가능합니다.

````python
# 변수 x와 y의 값을 바꿔봅시다.
# 그리고 결과를 print를 이용해 확인해봅시다.
````

````python
# 1) tmp 변수 활용
# x 재선언 시 기존 x 저장값은 덮어씌워짐
# 그러니 별도의 'tmp' 변수를 만들어서 백업해둔다는 개념
tmp = x
x = y
y = tmp
print(x, y)
````

````python
# 2) python 방식
# 변수 복수 선언 방식을 응용
# 사전에 변수 값을 확인할 필요가 있음
x, y = y, x
print(x, y)
````

### `None` 타입

파이썬에서는 값이 없음을 표현하기 위해 `None` 타입이 존재합니다.

````python
# None의 타입을 알아봅시다.
````

````python
print(type(None))
````

````python
# 변수에 저장해서 확인해봅시다.
# 변수 a에 None을 할당하고 출력해봅시다. (대분자 N)
````

````python
a = None
print(a)
# Null처럼 출력값이 없는 출력값
# None으로 출력되어도 실제로는 없는 값
````

* 참고) 함수의 return

````python
def my_sum(a, b):
    print(a + b)
    # print는 행동일 뿐, 결과값이 아님(print)
    # 함수의 결과값을 반환하는 return이 필수(output)
    return a + b


my_sum(3, 5)
````

````python
ans = my_sum(3, 5)
print(ans)
# return이 없으면 ans의 output이 없어 None이 출력됨
````

## 형변환(Type conversion, Typecasting)

파이썬에서 데이터타입은 서로 변환할 수 있습니다.

- 암시적 형변환
- 명시적 형변환

### 암시적 형변환(Implicit Type Conversion)

사용자가 의도하지 않았지만, 파이썬 내부적으로 자동으로 형변환 하는 경우입니다.
아래의 상황에서만 가능합니다.
* bool
* Numbers (int, float, complex)

````python
# boolean(T/F)과 integer는 더할 수 있을까요?
# True와 임의의 정수를 더해봅시다.
````

````python
True + 2
# True = 1, False = 0
# 문법에 관대한 Python에서만 가능한 일, 다른 자료형을 연산하는 것
````

````python
# int, float, complex를 각각 변수에 대입해봅시다.
# 변수 int_number 에 정수를 할당해봅시다.
# 변수 float_number 에 실수를 할당해봅시다.
# 변수 complex_number 에 복소수를 할당해봅시다.
````

````python
int_num = 11
float_num = 3.141592
# complex -> 복소수, i = j로 표현
complex_number = 2 + 3j
````

````python
# int와 float를 더해봅시다. 그리고 값을 출력해봅시다.
# 그 결과의 type은 무엇일까요?
````

````python
ans1 = int_num + float_num
print(ans1)
print(type(ans1))
````

````python
# int와 complex를 더해봅시다. 그리고 값을 출력해봅시다.
# 그 결과의 type은 무엇일까요?
````

````python
ans2 = int_num + complex_number
print(ans2)
print(type(ans2))

ans3 = float_num + complex_number
print(ans3)
print(type(ans3))
````

### 명시적 형변환(Explicit Type Conversion)
위의 상황을 제외하고는 모두 명시적으로 형변환을 해주어야합니다.

* string -> intger  : 형식에 맞는 숫자만 가능
* integer -> string : 모두 가능

암시적 형변환이 되는 모든 경우도 명시적으로 형변환이 가능합니다.

* `int()` : string, float를 int로 변환
* `float()` : string, int를 float로 변환
* `str()` : int, float, list, tuple, dictionary를 문자열로 변환

`list()`, `tuple()` 등은 다음 챕터에서 배울 예정입니다.

````python
# integer와 string 사이의 관계는 명시적으로 형변환을 해줘야만 합니다.
````

````python
# 정수와 문자열을 그냥 더해보고 오류를 확인해봅시다.
````

````python
5 + '6'
````

````python
# 정수를 문자열로 형변환하고 문자열과 더해봅시다.
````

````python
str(5) + '6'
````

````python
# 변수 a에 string 3을 할당하고 integer로 변환해봅시다.
````

````python
a = '3'
int(a)
# 정확한 숫자형일 경우 알아서 인식(띄어쓰기, 공백 등 절대 불가)
````

````python
# 변수 a에 string 3.5를 할당하고 float로 변환해봅시다.
````

````python
a = '3.5'
float(a)
````

````python
# string은 글자가 숫자일때만 형변환이 가능합니다.
# 변수 a에 문자열 'hi'를 할당하고 integer로 변환해봅시다.
````

````python
a = 'hi '
int(a)
````

````python
# float 3.5는 int로 변환이 가능합니다.
# 변수 a에 실수 3.5를 저장하고 integer로 변환해봅시다.
````

````python
a = 3.5
int(a)
# 버림
````
