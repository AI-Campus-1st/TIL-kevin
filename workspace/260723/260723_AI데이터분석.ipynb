```python
#1 '모음은 몇 개나 있을까'

word1=input('첫 번째 단어를 입력하세요.')
word2=input('두 번째 단어를 입력하세요.')
# print(word1,word2)

vowels=['a','e','i','o','u']
char1='_'.join(word1)
char2='_'.join(word2)
word1s=char1.split('_')
word2s=char2.split('_')
# print(word1s, word2s)

total1=0
for i in range(len(vowels)):
    num=word1s.count(vowels[i])
    total1+=num
print('%s'%word1+'에 들어간 모음의 수: %d'%total1)
    
total2=0
for i in range(len(vowels)):
    num=word2s.count(vowels[i])
    total2+=num
print('%s'%word2+'에 들어간 모음의 수: %d'%total2)
```

```python
#3 sorted()와 .sort()의 차이점

sample1=[8381,11379,47,1,99]
sample2=[8381,11379,47,1,99]
print(sample1)
print(sample2)

x=sample1.sort()
y=sorted(sample2)

print(sample1) # .sort()로 인해 sample1이 정렬되어 저장됨
print(x) # 하지만 해당 문장의 출력값은 None

print(sample2) # 원본은 그대로
print(y) # sorted 함수는 출력값이 정렬된 상태
```
