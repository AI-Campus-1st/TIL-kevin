```python
#1 복사가 잘 된 건가?
a = [1, 2, 3, 4, 5]
b = a # a의 list를 b도 동일하게 설정

a[2] = 5 # a의 list를 변경하여 

print(a)
print(b)
print(id(a)==id(b)) # 동일한 메모리 주소를 참조
print(a==b) # a == b -> True
```

```python
#2 평균 점수 구하기
scores = {
    'python': 80,
    'algorithm': 90,
    'django': 89,
    'web': 83,
} # => 85.5
# print(type(scores)) # dict

total=0
for key in scores.keys():
    value=scores[key]
    total+=value
print('평균 점수: %.2f'%(total/len(scores)))
```

```python
#3 혈액형 분류하기
# 여러 사람의 혈액형(A, B, AB, O)에 대한 정보가 담긴 bloods를 전달 받아, key는 혈액형, value는 사람 수인 dictionary를 만드시오.
bloods = [
    'A', 'B', 'A', 'O', 'AB', 'AB',
    'O', 'A', 'B', 'O', 'B', 'AB',
] # => {'A': 3, 'B': 3, 'O':  3, 'AB': 3}
# print(type(bloods)) # list

# 1)
dict={}
for i in bloods:
    # print(bloods)
    if i in dict:
        dict[i]+=1
    else:
        dict[i]=1
print(dict)
# 2)
dict={}
for i in bloods:
    cnt=bloods.count(i)
    dict[i]=cnt
print(dict)
# 3)
dict={}
for i in bloods:
    a=dict.get(i,0)
    dict[i]=a+1
print(dict)
```
