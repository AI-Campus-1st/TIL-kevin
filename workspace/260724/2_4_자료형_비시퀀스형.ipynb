## `set`

`set`은 순서가 없고 중복된 값이 없는 자료구조입니다.

* `set`은 수학에서의 집합과 동일하게 처리됩니다. 

* `set`은 중괄호`{}`를 통해 만들며, 순서가 없고 중복된 값이 없습니다.

* 빈 세트를 만들려면 `set()`을 사용해야 합니다. (`{}`로 사용 불가능)

* 활용 가능한 연산자는 차집합(`-`), 합집합(`|`), 교집합(`&`)입니다.

### 활용법
```python
{value1, value2, value3}
```

````python
# set 두개를 만들어서 연산자들을 활용해봅시다.
````

````python
set_a={1,3,5,7,9}
set_b={2,4,6,7,9}
````

* 차집합은 연산자 `-`를 사용합니다.

````python
# set_a와 set_b의 차집합을 구해봅시다.
set_a - set_b
set_b - set_a
````

* 합집합은 연산자 `|`를 사용합니다.

````python
# set_a와 set_b의 합집합을 구해봅시다.
set_a|set_b
````

* 교집합은 연산자 `&`을 사용합니다.

````python
# set_a와 set_b의 교집합을 구해봅시다.
set_a&set_b
````

````python
# set은 중복된 값이 있을 수 없습니다.
````

````python
set_c = {1, 1, 1}
print(set_c)
````

* `set`을 활용하면 `list`의 중복된 값을 손쉽게 제거할 수 있습니다.
* 단, `set`으로 변환하는 순간 순서를 보장할 수 없습니다.

````python
# set으로 중복된 값을 제거해봅시다.
# 정수 1, 2, 3, 1, 1, 2를 원소로 가지는 list를 만듭니다.
# 생성한 list를 set으로 형변환 합니다.
# 결과를 출력 해봅시다.
````

````python
nums=[1,2,3,1,1,2]
````

````python
set_nums=set(nums)
print(set_nums)
````

````python
# 다시 list로 바꿔서 확인해봅시다.
# list -> set -> list 바꾸면서 자연스럽게 중복이 제거됨
````

````python
nums = list(set_nums)
print(nums)
````

elem을 세트에 추가합니다.

## `dictionary`

`dictionary`는 `key`와 `value`가 쌍으로 이뤄져있습니다.


<center><img src="https://user-images.githubusercontent.com/18046097/61180427-1405d880-a651-11e9-94e1-1cc5c2a2ff34.png"></center> 

### 활용법

```python
{Key1:Value1, Key2:Value2, Key3:Value3, ...}
```

* `{}`를 통해 만들며, `dict()`로 만들 수 있습니다.
* `key`는 **변경 불가능(immutable)한 데이터**만 가능합니다. (immutable : string, integer, float, boolean, tuple, range)
* `value`는 `list`, `dictionary`를 포함한 모든 것이 가능합니다.

````python
# 비어있는 dictionary를 두가지 방법으로 만들어봅시다.
# {}와 dict()로 만들 수 있습니다.
# 두 변수의 타입을 출력해 봅시다.
````

````python
dict_a={}
dict_b=dict()
print(type(dict_a))
print(type(dict_b))
````

````python
my_dict = {[1, 2, 3]: 'my_str'}
# key로 list가 투입되었음 - dictionary는 immutable 자료형만 허용
````

````python
# dictionary에 중복된 key는 존재할 수가 없습니다.
````

````python
dict_a = {1: 1, 2: 2, 3: 3, 1: 4}
print(dict_a) # key 1이 덮어씌워짐
````

````python
# 지역번호가 담긴 전화번호부를 만들어봅시다.
# 변수 phone_book에 key를 지역명, value를 지역번호로 가지는 원소를 작성합니다.
# 예) 서울 - 02
````

````python
phone_book={
    '서울':'02',
    '강원':'033',
    '경기':'031',
    '인천':'032',
    '제주':'064',
    '부산':'051',
    '광주':'062',
    '대전':'042'
}
````

````python
# 위에서 작성한 phone_book이 가지고 있는 key 목록을 확인 해 봅시다.
# dictionary의 .keys() 메소드를 활용하여 key를 확인 해볼 수 있습니다.
````

````python
phone_book.keys()
````

````python
# 반복문을 활용해 key를 하나씩 출력해봅시다
````

````python
for key in phone_book.keys():
    print(key)
````

````python
# 위에서 작성한 phone_book이 가지고 있는 value 목록을 확인 해 봅시다.
# 딕셔너리의 .values() 메소드를 활용하여 value를 확인 해볼 수 있습니다.
````

````python
phone_book.values()
````

````python
# 반복문을 활용해 values를 하나씩 출력해봅시다.
````

````python
for value in phone_book.values():
    print(value)
````

````python
# 위에서 작성한 phone_book이 가지고 있는 key와 value 목록을 확인 해 봅시다.
# 딕셔너리의 .items() 메소드를 활용하여 key, value를 확인 해볼 수 있습니다.
````

````python
phone_book.items()
````

````python
# 반복문을 활용해 key, value 쌍을 하나씩 출력해봅시다.
````

````python
for items in phone_book.items():
    print(items)
    print(type(items)) # tuple은 순서가 있는 자료형
````

````python
# 참고) unpacking
````

````python
for items in phone_book.items():
    print(items) # tuple
    key,value=items
    print(key,value) # 두 항목을 분리한다는 의미의 unpacking
````

## 딕셔너리(Dictionary)

> 변경 가능하고(mutable), 순서가 없고(unordered), 순회 가능한(iterable)
>
> `Key: Value` 페어(pair)의 자료구조

데이터 구조로서의 딕셔너리(dictionary)와 조작법(method)

## 조회

### `.get(key[, default])`

key를 통해 value를 가져옵니다. 

절대로 KeyError가 발생하지 않습니다. default는 기본적으로 None입니다.

````python
# 변수명이 my_fruit인 딕셔너리를 만들어봅시다.
````

````python
my_fruit={'pineapple':3,'apple':5}
````

````python
# get 메서드 없이 딕셔너리 my_fruit의 key 'pineapple'의 value를 출력해봅시다.
````

````python
my_fruit['pineapple']
````

````python
# get 메서드로 딕셔너리 my_fruit의 key 'pineapple'의 value를 출력해봅시다.
````

````python
my_fruit['banana']
````

````python
my_fruit.get('banana') # None
# error 발생하지 않음
````

````python
my_fruit.get('banana',5) # 뒤의 옵션으로 key가 없을 때 value를 지정
````

````python
# get 메서드로 딕셔너리 my_fruit의 key 'apple'의 value를 출력해봅시다.
````

````python
# get 메서드로 딕셔너리 my_fruit의 key 'pineapple'의 value를 출력해봅시다.
# 단, key가 없다면 0을 반환하도록 해봅시다.
````

## 추가 및 삭제

### `.pop(key[, default])`

key가 딕셔너리에 있으면 제거하고 그 값을 돌려줍니다. 그렇지 않으면 default를 반환합니다.

default가 없는 상태에서 딕셔너리에 없으면 KeyError가 발생합니다.

````python
# 변수명이 my_fruit인 딕셔너리를 만들어봅시다.
````

````python
my_fruit={'apple':3,'grape':2}
````

````python
# pop 메서드로 딕셔너리 my_fruit의 key 'apple'을 제거해봅시다.
# 제거 후 딕셔너리 my_fruit를 출력해봅시다.
````

````python
result=my_fruit.pop('apple')
print(my_fruit)
print(result)
````

````python
# 제거하고자 하는 key가 딕셔너리에 없으면 KeyError가 발생합니다.
# 실행시켜 KeyError를 확인해봅시다.
my_fruit.pop('melon')
````

````python
# pop 메서드의 두번째 인자로 default 값을 설정 해 KeyError가 발생하지 않도록 할 수 있습니다.
# pop 메서드로 딕셔너리 my_dict의 key 'melon'을 제거하고 해당 key가 없다면 0을 반환하도록 해봅시다.
````

````python
result=my_fruit.pop('melon',0)
print(result)
````

### [연습] 딕셔너리 순회

> 혈액형 검사한 결과가 담긴 `blood_types`이 주어졌을때, 해당 딕셔너리를 순회하며, `key`와 `value`를 출력해보세요.

---

**[출력 예시]**
```
A형은 40명입니다.
B형은 11명입니다.
AB형은 4명입니다.
O형은 45명입니다.
```

````python
# 변수명이 blood_types인 딕셔너리를 만들어봅시다.
````

````python
blood_types={'A':40,'B':11,'AB':4,'O':45}
````

````python
# for문을 활용해봅시다.
# 딕셔너리 blood_types을 순회하며,
# 아래와 같은 문장이 출력되도록 하는 코드를 작성하세요.
"""
A형은 40명입니다.
B형은 11명입니다.
AB형은 4명입니다.
O형은 45명입니다.
"""
````

````python
for i in blood_types:
    print(i,blood_types[i]) # i=key
    value=blood_types[i]
    print('%s'%i+'형은 %d명입니다.\n'%value)
````

````python
# keys메서드를 활용해봅시다.
# blood_types.keys()을 순회하며,
# 아래와 같은 문장이 출력되도록 하는 코드를 작성하세요.
"""
A형은 40명입니다.
B형은 11명입니다.
AB형은 4명입니다.
O형은 45명입니다.
"""
````

````python
for key in blood_types.keys():
    print(key,blood_types[key])
    value=blood_types[key]
    print('%s형은 %d명입니다.'%(key,value)) # %(a,b,c,...)
````

````python
# items 메서드를 활용해봅시다.
# blood_types.items()을 순회하며,
# 아래와 같은 문장이 출력되도록 하는 코드를 작성하세요.
"""
A형은 40명입니다.
B형은 11명입니다.
AB형은 4명입니다.
O형은 45명입니다.
"""
````

````python
for item in blood_types.items():
    print(item)
    key,value=item
    key=item[0]
    value=item[1]
    print('%s형은 %d명입니다.'%(key,value))
````

### [실습] 딕셔너리 순회

> 혈액형 검사한 결과가 담긴 `blood_types`이 주어졌을때, 해당 검사에 참가한 사람들의 총합을 구해보세요.

---

**[출력 예시]**
```
검사에 참가한 사람은 총 100명입니다.
```

````python
# 변수명이 blood_types인 딕셔너리를 만들어봅시다.
````

````python
blood_types={'A':40,'B':11,'AB':4,'O':45}
````

````python
# for문을 통해 딕셔너리 blood_types를 순회하며,
# dict[key]로 접근하는 방법을 통해
# '검사에 참가한 사람은 총 100명입니다.' 라는 문장을 출력하는 코드를 작성하세요.
````

````python
total=0
for key in blood_types:
    print(key,blood_types[key])
    total+=blood_types[key]
print('검사에 참가한 인원은 총 %d명입니다.'%total)
````

````python
# 메서드 values를 활용해봅시다.
# for문을 통해 딕셔너리 blood_types를 순회하며,
# '검사에 참가한 사람은 총 100명입니다.' 라는 문장을 출력하는 코드를 작성하세요.
````

````python
total=0
for value in blood_types.values():
    print(value)
    total+=value
print('검사에 참가한 인원은 총 %d명입니다.'%total)
````

````python
# 빌트인 함수 sum을 사용해 
# '검사에 참가한 사람은 총 100명입니다.' 라는 문장을 출력하는 코드를 간단하게 작성해보세요.
````

````python
# sum([1,2,3,4]) -> 10
total=sum(blood_types.values())
print('검사에 참가한 인원은 총 %d명입니다.'%total)
````

### [응용] 딕셔너리 구축하기(counter)
> 리스트가 주어질 때, 각각의 요소의 개수를 value 값으로 갖는 딕셔너리를 만드세요.

---

**[출력 예시]**

{'great': 2, 'expectations': 1, 'the': 2, 'adventures': 2, 'of': 2, 'sherlock': 1, 'holmes': 1, 'gasby': 1, 'hamlet': 1, 'huckleberry': 1, 'fin': 1}


````python
# 변수명이 book_title인 리스트를 만들어봅시다.
````

````python
book_title =  ['great', 'expectations', 'the', 'adventures', 'of', 'sherlock', 'holmes', 
               'the', 'great', 'gasby', 'hamlet', 'adventures', 'of', 'huckleberry', 'fin']
````

````python
# 참고 - dict에서의 containment test

my_dict = {'a': '123', 'b': '456'}
if 'a' in my_dict: # key가 있는지 여부를 판단!
    print('있다!')
````

````python
# 1. dict[key]로 접근하는 방법을 통해 작성해보세요.
````

````python
counter={}
for title in book_title:
    print(title)
    if title in counter:
        counter[title]+=1   # title에 해당하는 value를 1 증가
    else:
        counter[title]=1    # title을 key, value는 1로 삼아 counter에 추가
print(counter)
````

````python
# 2. count 메서드를 활용해 작성해보세요.
````

````python
counter={}
for title in book_title:
    cnt=book_title.count(title)
    print(title,cnt)
    counter[title]=cnt
print(counter)
````

````python
# 3. get 메서드를 활용해 작성해보세요.
````

> `get(key[, default])`
* key 가 딕셔너리에 있는 경우 key 에 대응하는 값을 돌려주고, 그렇지 않으면 default 를 돌려줍니다. 

````python
counter={}
for title in book_title:
    a=counter.get(title,0)
    counter[title]=a+1
print(counter)
````

## 데이터의 분류
> `mutable` vs. `immutable`
[참고](https://www.geeksforgeeks.org/mutable-vs-immutable-objects-in-python/)

데이터는 크게 변경 가능한 것(`mutable`)들과 변경 불가능한 것(`immutable`)으로 나뉘며, python은 각각을 다르게 다룹니다.

### 변경 불가능한(`immutable`) 데이터

* 리터럴(literal)

    - 숫자(Number)
    - 글자(String)
    - 참/거짓(Bool)

* range()

* tuple()

* frozenset()



````python
print(range(10)[0]) # ?? 
print(range(10)[2]) # ?? 

# 결과는 어떻게 될까?
my_range = range(10)
my_range[2] = 2 # range is immutable(수정 불가능)
````

````python
# immutable 데이터의 복사는 어떻게 이루어질까?
num1 = 20
num2 = num1 # num2 -> 20
num2 = 10

print(num1)
print(num2)
````

## 리스트 복사

````python
# 변수 original_list에 원본 리스트를 저장합니다.
original_list = [1, 2, 3]
````

````python
# 변수 copy_list에 original_list를 저장하고 출력합니다.
copy_list = original_list
print(copy_list)
````

````python
# copy_list의 첫번째 값을 5로 바꾸고 original_list를 출력해봅시다.
copy_list[0] = 5
print(copy_list)
print(original_list)
````

````python
# copy_list, original_list 각각의 id 값을 == 연산자로 비교해봅시다.
print(id(copy_list) == id(original_list))
````

````python
copy_list = original_list.copy()
copy_list[0]=10
print(copy_list)
print(original_list)
print(id(copy_list),id(original_list))
````

````python
num1=[1,2,3,[1,2,3]]
num2=num1.copy()
num2[3][1]=100 # 3번 항목의 1번 항목
print(num1)
print(num2)
print(id(num1)==id(num2))

from copy import deepcopy
num2=deepcopy(num1) # 완전 복사
num2[3][1]=300
print(num1)
print(num2)
print(id(num1)==id(num2))
````
